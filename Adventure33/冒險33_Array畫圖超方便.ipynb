# 冒險 33 Array 畫圖超方便

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

只要有一個 list 或是 array，就可以畫圖。

```python
y = np.random.rand(50)
```

```python
y
```

```python
plt.plot(y)
```

用 array 畫函數。

先設 x 座標。

```python
x = np.linspace(0, 20, 100)
```

指定畫哪種函數。

```python
plt.plot(x, np.sin(x))
```

同時畫兩種函數也可以，還會自動用不同顏色標示。

```python
plt.plot(x, np.sin(x)+x) #圖示 legend
plt.plot(x, np.cos(x)-x)
```
