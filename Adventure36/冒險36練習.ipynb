# 冒險 36 練習

某大學中文系希望錄取文科成績好的學生。請從 grades.csv 的學測成績資料中找出國文、英文及社會科的平均成績大於等於 13 級分的學生姓名。

### 小題一 (10分)

1. 新增一欄位 "文科平均"。
2. 以 "文科平均" 進行排序，從排序中顯示出大於等於 13 級分的學生姓名。

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('https://bitly.com/gradescsv') # 在 Colab 中自網路讀取檔案較方便。
```

```python
df.head()
```

```python
df['文科平均'] = (df.國文 + df.英文 + df.社會) / 3
```

```python
df.head()
```

```python
df.sort_values(by = ['文科平均'], ascending = False).head(20)
```

### 小題二 (10 分)

使用「快篩法」(filter) 指令，僅列出 "文科平均" 大於等於 13 級分的學生。

參考教學：https://youtu.be/YPItfQ87qjM

作法一：不依 "文科平均" 排序

```python
df[df.文科平均 >= 13]
```

作法二：先依 "文科平均" 排序，再列出篩選。

```python
df2 = df.sort_values(by = ['文科平均'], ascending = False)
```

```python
df2[df2.文科平均 >= 13]
```
